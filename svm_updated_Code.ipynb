```python
data = pd.read_csv('/content/adult.data.txt')

test_data = pd.read_csv('/content/adult.test.txt')
```

```python

data = data.drop([' Zipcode', ' Education'], axis=1)

test_data = test_data.drop([' Zipcode', ' Education'], axis=1)
```

```python
data = data.drop([' Job'], axis = 1)

test_data = test_data.drop([' Job'], axis = 1)
```

```python
# prompt: add code to remove every line with '?' for employment or country

data = data[data[' Employment'] != ' ?']
data = data[data[' Country'] != ' ?']

test_data = test_data[test_data[' Employment'] != ' ?']
test_data = test_data[test_data[' Country'] != ' ?']
```

```python

print(data['Age'].unique())
print(data[' Income'].unique())
```

```python


data[' Gender'] = data[' Gender'].replace({' Female': 0, ' Male': 1})

test_data[' Gender'] = test_data[' Gender'].replace({' Female': 0, ' Male': 1})
```

```python
data[' Income'] = data[' Income'].replace({' <=50K': 0, ' >50K': 1})

test_data[' Income'] = test_data[' Income'].replace({' <=50K.': 0, ' >50K.': 1})
```

```python
test_data.head(30)
```

```python
columns = [col for col in data.columns if col != ' Income']
```

```python
x_train = data[columns]
y_train = data[' Income']

x_test = test_data[columns]
y_test = test_data[' Income']
```

```python

x_train = x_train.iloc[:1000]
y_train = y_train.iloc[:1000]

x_test = x_test.iloc[:1000]
y_test = y_test.iloc[:1000]
```

```python

x_train = x_train.drop([' Employment', ' Marital_Status', ' Relationship', ' Race', ' Country'], axis=1)

x_test = x_test.drop([' Employment', ' Marital_Status', ' Relationship', ' Race', ' Country'], axis=1)
```

```python
x_test.head(30)
```

```python
#feature scaling of data
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x_train)

x_test_scaled = scaler.transform(x_test)
```

```python
x_tensor = torch.FloatTensor(x_scaled)
y_tensor = torch.FloatTensor(y_train.values)
x_test_tensor = torch.FloatTensor(x_test_scaled)
y_test_tensor = torch.FloatTensor(y_test.values)
```

```python
#defines one layer neural network

class SVM(nn.Module):
    def __init__(self, input_dim):
        super(SVM, self).__init__()
        self.linear = nn.Linear(input_dim, 1)

    def forward(self, x):
        output = self.linear(x)
        return output
```

```python
learning_rate = 0.01 #step size in each iteration of SGD
lambda_param = 0.01 # controls strenght of regularization
num_epochs = 100
batch_size = 64
```

```python
input_dim = x_tensor.shape[1]
model = SVM(input_dim)
```

```python
train_dataset = TensorDataset(x_tensor, y_tensor.view(-1, 1))
train_loader = DataLoader(dataset=train_dataset, batch_size=batch_size, shuffle=True)
```

```python
optimizer = optim.SGD(model.parameters(), lr=learning_rate)
```

```python
#loss function common in classification, maximizes distance between boundary and nearest datapoints
def hinge_loss(outputs, labels):
    # Convert labels from [0, 1] to [-1, 1]
    labels = 2 * labels - 1
    # Calculate hinge loss
    return torch.mean(torch.clamp(1 - outputs * labels, min=0))
```

```python
model.train()
for epoch in range(num_epochs):
    total_loss = 0
    for inputs, labels in train_loader:
        # Forward pass
        outputs = model(inputs)

        # Compute hinge loss
        loss = hinge_loss(outputs, labels)

        # Add L2 regularization term (weight decay, based on squared sum of weights)
        l2_reg = torch.tensor(0.)
        for param in model.parameters():
            l2_reg += torch.norm(param, p=2)
        loss += lambda_param * l2_reg

        # Compute gradients and optimize model accordingly
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}')
```

```python
model.eval()
with torch.no_grad():
    y_pred_tensor = model(x_test_tensor)
    # Convert outputs to binary predictions
    y_pred = (y_pred_tensor > 0).float().numpy().flatten() #converts predictions to 0 if false and 1 if true

accuracy = accuracy_score(y_test, y_pred)
print(f"\nAccuracy: {accuracy:.4f}")
```

```python
precision = precision_score(y_test, y_pred)
print(f"\nPrecision: {precision:.4f}")
```

```python
recall = recall_score(y_test, y_pred)
print(f"Recall: {recall:.4f}")
```

```python
f1 = f1_score(y_test, y_pred)
print(f"F1 Score: {f1:.4f}")
```
